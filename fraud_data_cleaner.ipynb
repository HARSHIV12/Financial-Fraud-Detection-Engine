```python
query = """
SELECT 
    t1.trans_num,
    t1.cc_num,
    t1.trans_date_trans_time,
    t1.amt,
    t2.trans_date_trans_time AS previous_time,
    ROUND((JULIANDAY(t1.trans_date_trans_time) - JULIANDAY(t2.trans_date_trans_time)) * 24 * 60, 2) AS minutes_difference
FROM transactions_cleaned t1
JOIN (
    SELECT 
        trans_num, 
        cc_num, 
        trans_date_trans_time,
        LAG(trans_date_trans_time) OVER (PARTITION BY cc_num ORDER BY trans_date_trans_time) AS previous_time
    FROM transactions_cleaned
) t2
ON t1.trans_num = t2.trans_num
WHERE 
    previous_time IS NOT NULL
    AND ROUND((JULIANDAY(t1.trans_date_trans_time) - JULIANDAY(t2.previous_time)) * 24 * 60, 2) < 2;
"""
```

```python
pd.read_sql_query(query, conn)
```

```python
cursor.execute("""
CREATE VIEW IF NOT EXISTS trans_with_location AS
SELECT 
    t.trans_num,
    t.cc_num,
    t.trans_date_trans_time,
    t.unix_time,
    m.merchant,
    m.merch_lat,
    m.merch_long
FROM transactions_cleaned t
JOIN merchants m ON t.merchant = m.merchant
""")
conn.commit()
```

```python
query2 = """
WITH ordered_trans AS (
    SELECT *,
        LAG(merch_lat) OVER (PARTITION BY cc_num ORDER BY unix_time) AS prev_lat,
        LAG(merch_long) OVER (PARTITION BY cc_num ORDER BY unix_time) AS prev_long,
        LAG(unix_time) OVER (PARTITION BY cc_num ORDER BY unix_time) AS prev_unix_time
    FROM trans_with_location
),
distances AS (
    SELECT *,
        ( ( (merch_lat - prev_lat)*(merch_lat - prev_lat) + 
            (merch_long - prev_long)*(merch_long - prev_long) ) * 111 * 111 ) AS distance_km,
        (unix_time - prev_unix_time) / 3600.0 AS time_diff_hr
    FROM ordered_trans
    WHERE prev_lat IS NOT NULL AND prev_long IS NOT NULL
)
SELECT * FROM distances
WHERE distance_km / time_diff_hr > 1000;
"""
```

```python
pd.read_sql_query(query2, conn)
```

```python
query3 = """
WITH avg_spend AS (
    SELECT 
        cc_num,
        AVG(amt) AS avg_amt
    FROM transactions_cleaned
    GROUP BY cc_num
)
SELECT 
    t.trans_num,
    t.cc_num,
    t.amt,
    a.avg_amt
FROM transactions_cleaned t
JOIN avg_spend a ON t.cc_num = a.cc_num
WHERE t.amt > 5 * a.avg_amt;
"""
```

```python
pd.read_sql_query(query3, conn)
```

```python
query4 = """
WITH category_count AS (
    SELECT cc_num, category, COUNT(*) AS cnt
    FROM transactions_cleaned t
    JOIN merchants m ON t.merchant = m.merchant
    GROUP BY cc_num, category
),
main_category AS (
    SELECT cc_num, category
    FROM (
        SELECT cc_num, category, RANK() OVER (PARTITION BY cc_num ORDER BY cnt DESC) AS rk
        FROM category_count
    ) WHERE rk = 1
)
SELECT t.trans_num, t.cc_num, m.category, mc.category AS usual_category
FROM transactions_cleaned t
JOIN merchants m ON t.merchant = m.merchant
JOIN main_category mc ON t.cc_num = mc.cc_num
WHERE m.category != mc.category;
"""
```

```python
pd.read_sql_query(query4, conn)
```
